```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sb
```

```python
df = pd.read_csv(r"D:\Desktop\jhk\elevate labs\task 5\train.csv")
#1
print(df.head())
#2
print("\n--- Info ---")
df.info()
#3
print("\n--- Describe ---")
print(df.describe())
```

# histograms



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1) 
sb.histplot(df['Age'].dropna(), kde=True, bins=30) 
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')

plt.subplot(1, 2, 2) 
sb.histplot(df['Fare'], kde=True, bins=30)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Count')

#plt.tight_layout() 
plt.show()
```

# Boxplots


```python
plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1) 
sb.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2) 
sb.boxplot(x='Sex', y='Age', data=df)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()

#plt.subplot(1, 3, 3) 
plt.figure(figsize=(8, 6))
sb.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

# Scatterplots 


```python
plt.figure(figsize=(10, 12))

plt.subplot(2, 1, 1)
sb.scatterplot(x='Age', y='Pclass', hue='Survived', data=df, alpha=0.8)
plt.title('Age vs. Pclass (colored by survival)')
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.legend(title='Survived') 


plt.subplot(2, 1, 2)
sb.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.8)
plt.title('Age vs. Fare (colored by survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived') 

plt.tight_layout()
plt.show()
```

# pairplot


```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']

sb.pairplot(df[numerical_cols].dropna(), hue='Survived') 
plt.suptitle('Pair Plot of Numerical Features by Survival', y=1.02) 
plt.show()
```

# heatmap


```python
correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
